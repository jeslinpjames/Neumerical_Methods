```python
def function(x):
    return x**3 - 4*x - 9
def false_position_method(e):
    j = 0
    while(function(j)*function(j+1)>0):
        j+=1
    a = j
    b = j + 1
    c = (a*function(b) - b*function(a))/(function(b) - function(a))
    iteration = 0
    while((b-c)>=e):
        if(function(c)<0):
            a = c 
        else:
            b = c
        c = (a*function(b) - b*function(a))/(function(b) - function(a))
        iteration+=1
        print("Iteration: ",iteration)
        print("a: ",a)
        print("b: ",b)
        print("c: ",c)
        print("f(c): ",function(c))
    return c
c=false_position_method(0.0005)
print(c)
```
