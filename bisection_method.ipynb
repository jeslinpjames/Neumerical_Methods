```python
a = 2
b = 3
prev_cn = 0  

for iteration in range(1, 500):
    cn = (a + b) / 2
    fn = cn**3 - 4*cn - 9

    print(f"Iteration {iteration}:")
    print(f"  cn: {cn}")
    print(f"  fn: {fn}")
    print(f"  a: {a}")
    print(f"  b: {b}")

    if fn > 0:
        b = cn
    else:
        a = cn

    if abs(cn - prev_cn) <= 0.00005:
        break

    prev_cn = cn
```

```python
def bisection_method(fn, tolerance):
    prev_cn =0
    iteration = 1 
    j = 0
    while(function(j)*function(j+1)>0):
        j+=1
    a = j
    b = j + 1
    while True:
        cn = (a + b) / 2
        fn_val = fn(cn)
        print(f"Iteration {iteration}:")
        print(f"  cn: {cn}")
        print(f"  fn: {fn_val}")
        print(f"  a: {a}")
        print(f"  b: {b}")
        if fn_val>0:
            b = cn
        else:
            a = cn
        if abs(cn - prev_cn) <= tolerance:
            print("Converged at iteration =", iteration)
            break
        prev_cn = cn
        iteration+=1
    return cn
        
def function(x):
    return x**3 - 4*x - 9

result = bisection_method(function,0.00005)
print(f"\nApproximate root: {result}")
```
